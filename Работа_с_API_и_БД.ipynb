# Работа с API и БД

Работа с API

```python
import requests
import pandas as pd
```

```python
# Информацию обо всех книгах
URL = 'https://www.anapioficeandfire.com/api/books'
response = requests.get(URL)
book = response.json()
df_book = pd.DataFrame(book)
df_book.head()
```

```python
# Информацию обо всех домах Вестероса
URL = 'https://www.anapioficeandfire.com/api/houses'
response = requests.get(URL, params={'region': "The Westerlands"})
houses = response.json()
df_houses = pd.DataFrame(houses)
df_houses
```

```python
# Информацию обо всех домах Вестероса, у которых есть девиз
URL = 'https://www.anapioficeandfire.com/api/houses'
response = requests.get(URL, params={'region': "The Westerlands"})
houses = response.json()

houses_motto = [house for house in houses if house.get('words')]

df_houses_motto = pd.DataFrame(houses_motto)
df_houses_motto
```
