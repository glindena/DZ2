```python
df_pd_h = pd.read_csv('Housing.csv')
```

```python
## "price" - так как минимальное значение равно 1750000, а максимальное значение равно 13300000, то с точки зрения потребления памяти оптимальным типом данных будет int32
## "area" - так как минимальное значение равно 1650, а максимальное значение равно 16200, то с точки зрения потребления памяти оптимальным типом данных будет int16
## "bedrooms" - так как минимальное значение равно 1, а максимальное значение равно 6, то с точки зрения потребления памяти оптимальным типом данных будет int8
## "bathrooms" так как минимальное значение равно 1, а максимальное значение равно 4, то с точки зрения потребления памяти оптимальным типом данных будет int8
## "stories" - так как минимальное значение равно 1, а максимальное значение равно 4, то с точки зрения потребления памяти оптимальным типом данных будет int8
## "mainroad", "guestroom", "basement", "hotwaterheating", "airconditioning", "prefarea" и "furnishingstatus" - так как эти столбцы содержат всего по две фразы "yes" и "no",
## то с точки зрения потребления памяти оптимальным типом данных будет category
## "parking" - так как минимальное значение равно 0, а максимальное значение равно 3, то с точки зрения потребления памяти оптимальным типом данных будет int8
```

```python
df_pd_h.info()
```

```python
df_pd_h = df_pd_h.fillna(0)
df_pd_h['price'] = df_pd_h['price'].astype(np.int32)
df_pd_h['area'] = df_pd_h['area'].astype(np.int16)
df_pd_h['bedrooms'] = df_pd_h['bedrooms'].astype(np.int8)
df_pd_h['bathrooms'] = df_pd_h['bathrooms'].astype(np.int8)
df_pd_h['stories'] = df_pd_h['stories'].astype(np.int8)
df_pd_h['mainroad'] = df_pd_h['mainroad'].astype("category")
df_pd_h['guestroom'] = df_pd_h['guestroom'].astype("category")
df_pd_h['basement'] = df_pd_h['basement'].astype("category")
df_pd_h['hotwaterheating'] = df_pd_h['hotwaterheating'].astype("category")
df_pd_h['airconditioning'] = df_pd_h['airconditioning'].astype("category")
df_pd_h['parking'] = df_pd_h['parking'].astype(np.int8)
df_pd_h['prefarea'] = df_pd_h['prefarea'].astype("category")
df_pd_h['furnishingstatus'] = df_pd_h['furnishingstatus'].astype("category")
df_pd_h.info()
```
