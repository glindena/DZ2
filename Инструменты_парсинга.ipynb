# Инструменты парсинга

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
```

```python
base_url = 'https://lifehacker.ru' # Базовая часть ссылки
mask = 'https://' # Маска для фильтрации рекламных ссылок
parsed_urls = []

for page_num in range(1, 11):
    url = f'{base_url}/topics/technology/?page={page_num}' # Ссылка для пагинации
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    raw_items = soup.find_all('a', class_='lh-small-article-card__link')
    parsed_urls += [item.get('href')
      for item in raw_items
        if item.get('href') and mask not in item.get('href')] # Фильтруем ссылки по маске
```

```python
result = []

for url in tqdm(parsed_urls):
    article = {}

    article_url = f'{base_url}{url}' # Подставляем часть ссылки, ведущую на материал
    response = requests.get(article_url)
    soup = BeautifulSoup(response.text, 'lxml')

    article['Заголовок'] = soup.find('h1', class_='article-card__title').text # Получаем текст заголовка
    article['Текст'] = soup.find('article', class_='single-article__post-content').text # Получаем весь текст статьи

    result.append(article)
```

```python
pd.set_option('display.max_colwidth', 400)
data = pd.DataFrame(result)
data.head()
```
