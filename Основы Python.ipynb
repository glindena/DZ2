```python
class  Chocolate:
  def __init__(self, price, additive):
    self.price = price
    self.additive = additive
    self.price = float (self.price)
    self.additive = str (self.additive)

  def show_my_chocolate(self):
     if self.additive != '' and self.price >0:
        return f"Шоколадный батончик, добавка: {self.additive}"
     elif self.price <0:
        raise ValueError('Цена батончика не может быть отрицательной!')
     else:
        return f"Обычный шоколадный батончик"
  def __str__(self):
     return f"{self.price} {self.additive}"

Snikers = Chocolate('55.5',"")
print (Snikers.show_my_chocolate())

print ("Шоколадный батончик,", Snikers.price,"рублей")

Mars = Chocolate('3',"карамель")
print (Mars.show_my_chocolate())

Baunty = Chocolate('-1',"кокос")
print (Baunty.show_my_chocolate())
```
