```python
class  Chocolate:
  def __init__(self, price, additive):
    self.price = price
    self.additive = additive
    self.price = float (self.price)
    self.additive = str (self.additive)

    if self.price < 0:
     raise ValueError('Цена батончика не может быть отрицательной!')

  def show_my_chocolate(self):
     if self.additive != '':
        return f'Шоколадный батончик, добавка: {self.additive}'
     else:
        return f'Обычный шоколадный батончик'

  def __str__(self):
     return f'Шоколадный батончик, {self.price} рублей'

##Пример 1
Snikers = Chocolate('60','')
print (1, Snikers.show_my_chocolate())
print(Snikers, '\n')

##Пример 2
Mars = Chocolate('3',"карамель")
print (2, Mars.show_my_chocolate(), '\n')

##Пример 3
Twix = Chocolate('45',"печенье")
print (3, Twix.show_my_chocolate())
print(Twix, '\n')

##Пример 4
Baunty = Chocolate('-1',"кокос")
print (4, Baunty.show_my_chocolate())
```
